```python
import pandas as pd
df=pd.read_csv("/mnt/sde1/geth/nft_analyse_v1/data/giveBirth/csv/givebirth_withRevenue.csv")

totalCount=sum(df["rewardCount"])
totalRevenue=0

totalCost=0

for index,row in df.iterrows():
    ethPrice=float(row["ethPrice"])
    callReward=float(row["callReward"])/1e18
    
    totalRevenue+=(ethPrice*callReward)
    
    transactionFee=float(row["transactionFee"])/1e18
    coinbase_transfer=float(row["coinbase_transfer"])/1e18
    totalCost+=((transactionFee+coinbase_transfer)*ethPrice)
    
print("giveBirth平均每次调用收益: ",totalRevenue/totalCount)

print("每笔交易的成本： ",totalCost/len(df))
```

```python
import pandas as pd
df=pd.read_csv("/mnt/sde1/geth/nft_analyse_v1/data/giveBirth/csv/givebirth_withRevenue.csv")

totalRevenue=0
totalCount=0
totalFee=0

for index,row in df.iterrows():
    ethPrice=float(row["ethPrice"])
    callReward=float(row["callReward"])/1e18
    rewardCount=row["rewardCount"] 
    transactionFee=float(row["transactionFee"])/1e18
    coinbase_transfer=float(row["coinbase_transfer"])/1e18
    
    if coinbase_transfer==0:
        continue
    
    totalFee+=(transactionFee*ethPrice)
    totalRevenue+=(ethPrice*callReward)
    totalCount+=rewardCount

print("totalRevenue",totalRevenue)
print("totalCount",totalCount)
print("totalFee",totalFee)
```
